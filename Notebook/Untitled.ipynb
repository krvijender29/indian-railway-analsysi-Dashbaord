**Data Cleaning &  Data Quality Check**

```python
# Misiing Values 
print("Missing Values in Train Dataset")

train_missing = train_df.isnull().sum().sort_values(ascending = False)

tarin_missing = pd.DataFrame({
    "missing Values": train_missing,
    "Percenatge" : round(train_missing / len(train_df)*100,2)
})
train_missing
```

```python
#missing Values in Stations Dataset
print("Missing Values in Station Dataset")

station_missing = station_df.isnull().sum().sort_values(ascending=False)

station_missing = pd.DataFrame({
    "Missing Values": station_missing,
    "Percentage": round((station_missing / len(station_df)) * 100, 2)
})

station_missing
```

```python
#missing Values in Schedule Dataset
print("Missing Values in Schedule Dataset")

schedule_missing = schedule_df.isnull().sum().sort_values(ascending=False)

schedule_missing = pd.DataFrame({
    "Missing Values": schedule_missing,
    "Percentage": round((schedule_missing / len(schedule_df)) * 100, 2)
})

schedule_missing
```

```python
print("Duplicate Records")
# In Station and train Dataset one column is list so duplicate function not work at list so we convert the list to tuple in both dataset
train_df["geometry.coordinates"] = train_df["geometry.coordinates"].apply(
    lambda x: tuple(map(tuple, x)) if isinstance(x, list) else x
)

station_df["geometry.coordinates"] = station_df["geometry.coordinates"].apply(
    lambda x: tuple(x) if isinstance(x, list) else x
)
print(f"Train Dataset    : {train_df.duplicated().sum()}")
print(f"Station Dataset  : {station_df.duplicated().sum()}")
print(f"Schedule Dataset : {schedule_df.duplicated().sum()}")
```

```python
# Percenatge of Missing Values
def missing_percentage(df):

    missing = pd.DataFrame({
        "Missing %": round(df.isnull().mean() * 100,2)
    })

    return missing[missing["Missing %"]>0].sort_values("Missing %",ascending=False)

train_missing_per=missing_percentage(train_df)
station_missing_per=missing_percentage(station_df)
schedule_missing_per=missing_percentage(schedule_df)
print("Train Dataset Missing Values % ")
print("-"*50)
print(train_missing_per)
print("-"*50)
print("Station Dataset Missing Values % ")
print("-"*50)
print(station_missing_per)
print("-"*50)
print("Schedule Dataset Missing Values %  ")
print("-"*50)
print(schedule_missing_per)
```
